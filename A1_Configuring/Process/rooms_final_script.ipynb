```python
def room(Name, Area_function):
    Name = str(Name)
    x = list(range(0,400))
    y = list(range(0,400))
    Grid_size = 2
    Area_function = Area_function
    dispatch_relation = 0.33

    Area_function_high = Area_function
    Area_function_low = Area_function
    Original_Value = Area_function
    final_pairs = []
    values = []



    for i in (list(range(180))):
        values.append((Grid_size**2)*i)

    def takeClosest(Values, Area_function):
        newlst = []
        for i in values:
            newlst.append(i - Area_function)
        lstt = [abs(ele) for ele in newlst]
        _A_ = values[lstt.index(min(lstt))]
        return _A_

    def x_and_yRange (x, y, _Area_function, Grid_size):
        x_coordinate = []
        y_coordinate = []
        pairs = []

        for i in x:
            for j in y:
                if (Grid_size*i)*(Grid_size*j) == _Area_function:
                    x_coordinate.append(i)
                    y_coordinate.append(j)
                else:
                    pass
        
        for t in range(len(x_coordinate)):
            if t < ((1/2)*(len(x_coordinate))):
                pairs.append(x_coordinate[t])
                pairs.append(y_coordinate[t])
            else:
                pass


        return pairs, x_coordinate, y_coordinate


    _Area_function = takeClosest(values, Area_function)

    outcome = x_and_yRange (x, y, _Area_function, Grid_size)


    pair_size = outcome[0]

    x_size = outcome[1]
    x_cells = x_size[:int((len(x_size))/2)]


    y_size = outcome[2]
    y_cells = y_size[:int((len(y_size))/2)]




    pairs = [pair_size[i:i + 2] for i in range(0, len(pair_size), 2)]




    for i in pairs:
        if i[0]/5 != 1 and float(i[0]/i[1]) > dispatch_relation :
            final_pairs.append(i)
        else :
            pass
    

    while len(final_pairs) < 2 :
        Area_function_high = Area_function_high + 1
        Area_function_low = Area_function_low - 1
        

        for i in (list(range(180))):
            values.append((Grid_size**2)*i)

        _Area_function = takeClosest(values, Area_function_high)
        _area_function_low = takeClosest(values, Area_function_low)

        outcome_low = x_and_yRange(x, y, _area_function_low, Grid_size)
        outcome = x_and_yRange (x, y, _Area_function, Grid_size)

        pair_size_low = outcome_low[0]
        pair_size = outcome[0]

        x_size_low = outcome_low[1]
        x_size = outcome[1]
        x_cells_low = x_size_low[:int((len(x_size_low))/2)]
        x_cells = x_size[:int((len(x_size))/2)]

        y_size_low = outcome_low[2]
        y_size = outcome[2]
        y_cells_low = y_size_low[:int((len(y_size_low))/2)]
        y_cells = y_size[:int((len(y_size))/2)]


        pairs_low = [pair_size_low[i:i+2] for i in range (0, len(pair_size_low),2)]
        pairs = [pair_size[i:i + 2] for i in range(0, len(pair_size), 2)]

        for i in pairs_low:
            if i[0]/5 != 1 and float(i[0]/i[1]) > dispatch_relation :
                final_pairs.append(i)
            else :
                pass
        if Area_function_low == Original_Value - 15:
            break


        for i in pairs:
            if i[0]/5 != 1 and float(i[0]/i[1]) > dispatch_relation :
                final_pairs.append(i)
            else :
                pass
        if Area_function_high == 15 + Original_Value:
            break
    
    return print(Name), Name, _Area_function, final_pairs
    
```
